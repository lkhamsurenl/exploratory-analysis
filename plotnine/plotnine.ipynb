```python
# this is computing custom summary stats for given by fun arguments that 
# happened to be along the y axis
(diamonds
 .pipe(ggplot)
 +stat_summary(
    mapping = aes(x = "cut", y = "depth"),
    fun_ymin=np.min,
    fun_ymax=np.max,
    fun_y=np.median
  )
)
```

```python

from matplotlib import rc
rc('text', usetex=True)

df = pd.DataFrame({"x": np.random.uniform(size=10),
                   "y": np.random.uniform(size=10)})

(df
 .pipe(ggplot)
 +aes("x", "y")
 +geom_point()
 +labs(
    x="$\\sum_{i = 1}^n{x_i^2}$",
    y="$\\alpha + \\beta + \\frac{\\delta}{\\theta}$"
 )
)
```

```python
best_in_class = (mpg
.sort_values(by="hwy", ascending=False)
.groupby("class")
.first()
)

(mpg
 .pipe(ggplot)
 +aes("displ", "hwy")
 +geom_point(aes(color = "class"))
 +geom_text(
        data = best_in_class, 
        mapping = aes(label = "model")
    )
)
```
